```python
import pandas as pd
```

```python
df=pd.read_csv('IRIS.csv')
df.head(2)
```

```python
df.info()
```

```python
#null value treatment
df.isna().sum()
```

```python
#separate input and output 
x=df.drop(labels=['species'],axis=1) #inputes
y=df[['species']] #outputs
```

```python
#preprocesssing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform= 'pandas')
x_pre = scaler.fit_transform(x)
```

```python
import pickle #pickle used to bridge of frontend and backend
with open('scaler.pkl','wb')as file1:
    pickle.dump(scaler,file1) 
```

```python
#train,test,split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_pre,y,test_size=0.33,random_state=21)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytest.value_counts()
```

```python
#model building
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
model.score(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
#to load any pickle file
with open('model.pkl','wb')as file2:
    pickle.dump(model,file2)
```
